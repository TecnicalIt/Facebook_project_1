### overview. two columns null values accept. it's likes and another share.I have handling null values.
সংক্ষিপ্ত বিবরণ। দুটি কলাম শূন্য মান গ্রহণ করে। এটি লাইক এবং আরেকটি শেয়ার।আমি শূন্য মান গুলি পরিচালনা করেছি। 


```python
null_likes_mean = df.like.mean()
```

```python
null_likes_mean
```

```python
null_likes_median = df.like.median()
null_likes_median
```

```python
df["like"] = df.like.fillna(null_likes_median)
```

```python
df["like"].isnull().any()
```

```python
null_share_mean = df.share.mean()
```

```python
null_share_mean
```

```python
null_share_median = df.share.median()
null_share_median
```

```python
df["share"] = df.share.fillna(null_share_median)
```

```python
df["Paid"] = df.Paid.fillna(1.0)
```

```python
df["share"].isnull().any()
```

```python
x = df.drop("Total Interactions",axis=1)
```

```python
y = df["Total Interactions"]
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(x.corr().T,annot=True,cmap="RdYlGn")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
for column in x.columns:
    if x[column].dtype == np.number:
        continue
    x[column] =  LabelEncoder().fit_transform(x[column])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
